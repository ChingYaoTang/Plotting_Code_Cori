# Phase plot for clumps

```python
output_dir_pp = "./PlotForResult/clump_phase_plot.png"
field_x = "density"
field_x_label = r"$\mathrm{Density \ [g/cm^3]}$"
field_y = "temperature"
field_y_label = r"$\mathrm{Temperature \ [K]}$"
field_z = "H2_fraction"
field_z_label = r"$\mathrm{H_2 \ fraction}$"
xticks = [1e-18,1e-17,1e-16]
yticks = [200,400,600,800,1000,1200]
ncol = 3
nrow = 2
figsize_x = 30
figsize_y = 30
tick_size = 14
label_size = 20
title_size = 16
```

```python
#output_dir  = ["/test2203281800", "/test2203281500",
#               "/test2110121500", "/test2110121400"]

output_dir  = ["/testM4C3S","/testM4C4D","/testM8C3S","/testM8C4S","/testM8C4D","/testM8C4D"]

ds_clumps   = [read_clumps_info(cscratch+output_dir[i]+"/clump_info.h5") for i in range(len(output_dir))]
```

```python
which_3rd   = [122,148,160,120,122,122] 
which_dump  = [8,17,16,12,4,4]
which_leave = [0,0,0,0,3,1]
#Clumps  = [master_clump.leaves[i] for i in range(4)]
Clumps  = [ds_clumps[i].tree.children[which_leave[i]] for i in range(len(output_dir))]
Files   = ["{}/{}/data-3rdStep-{}/DD{:0>4d}/DD{:0>4d}".format(cscratch,output_dir[i], which_3rd[i],
                                                           which_dump[i], which_dump[i]) for i in range(len(output_dir))]
model_name = ["M4C3S-a", "M4C4D-a", "M8C3S-a", "M8C4S-a", "M8C4D-a", "M8C4D-b"]
#model_name = ["{:<7.3f} ".format(float(Clumps[i]["grid","cell_mass"].sum().in_units("Msun")))+r"$M_\odot$" for i in range(len(output_dir))]
```

```python
def clump_phase_diagram(Clumps, Files, output_dir, ncol, nrow, figsize_x, figsize_y, label_size, tick_size, model_name,
                        field_x, field_x_label,
                        field_y, field_y_label,
                        field_z, field_z_label,
                        title_size, xticks, yticks,
                        weight_field=None, colormap='jet'):
    
    if (len(Clumps) != ncol*nrow):
        print("Length of Files does not equal to the total number of plots.\n")
        return 0
    
    
    # Find the upper and lower limits of each field
    min_field_x = Clumps[0]["grid", field_x].min()
    max_field_x = Clumps[0]["grid", field_x].max()
    min_field_y = Clumps[0]["grid", field_y].min()
    max_field_y = Clumps[0]["grid", field_y].max()
    min_field_z = Clumps[0]["grid", field_z].min()
    max_field_z = Clumps[0]["grid", field_z].max()
    for i in range(1, len(Clumps)):
        min_field_x_tmp      = Clumps[i]["grid", field_x].min()
        max_field_x_tmp      = Clumps[i]["grid", field_x].max()
        min_field_x          = np.min([min_field_x_tmp, min_field_x])
        max_field_x          = np.max([max_field_x_tmp, max_field_x])
        
        min_field_y_tmp      = Clumps[i]["grid", field_y].min()
        max_field_y_tmp      = Clumps[i]["grid", field_y].max()
        min_field_y          = np.min([min_field_y_tmp, min_field_y])
        max_field_y          = np.max([max_field_y_tmp, max_field_y])
        
        min_field_z_tmp      = Clumps[i]["grid", field_z].min()
        max_field_z_tmp      = Clumps[i]["grid", field_z].max()
        min_field_z          = np.min([min_field_z_tmp, min_field_z])
        max_field_z          = np.max([max_field_z_tmp, max_field_z])
        
    # Create a blank figure
    fig = plt.figure(figsize=(figsize_x, figsize_y))
    # See http://matplotlib.org/mpl_toolkits/axes_grid/api/axes_grid_api.html
    grid = AxesGrid(fig, rect=(0.08,0.08,0.8,0.88),
                    nrows_ncols = (nrow, ncol),
                    axes_pad = 0.4,
                    label_mode = "L",
                    share_all = True,
                    cbar_location="right",
                    cbar_mode="single",
                    cbar_size="3%",
                    cbar_pad="2%",
                    aspect = False
                   )
    
    ds_list = []
    for file in Files:
        ds_list.append(yt.load( file ))
        
    for i, ds in enumerate(ds_list):
        clump = Clumps_Info(Clumps[i])
        print("clump max density position = ", [clump.max_density_pos_x.in_units("cm"),
                                                clump.max_density_pos_y.in_units("cm"),
                                                clump.max_density_pos_z.in_units("cm")])
        # Load the data and create a single plot
        if i!=3:
            sp = ds.sphere((float(clump.max_density_pos_x.in_units("code_length")),
                            float(clump.max_density_pos_y.in_units("code_length")),
                            float(clump.max_density_pos_z.in_units("code_length"))),
                           (float(clump.avg_radius.in_units("pc"))*1.0, "pc"))
        else:
            sp = ds.sphere((2.2661135889901374/3, 2.1020510667103083/3, 0.10693360827167422/3),
                           (0.11, "pc"))
        
        print("sphere center position     = ",sp)
        p  = yt.PhasePlot( sp, field_x, field_y, [field_z], weight_field=weight_field)
        
        # Ensure the axes and colorbar limits match for all plots
        p.set_xlim( min_field_x, max_field_x )
        p.set_ylim( min_field_y, max_field_y )
        p.set_zlim( field_z, min_field_z, max_field_z )
        p.set_cmap(field=field_z, cmap=colormap)
        p.set_log(field_y, False)
        #p.set_log(field_z, False)
        #p.set_figure_size(4)
        
        # This forces the ProjectionPlot to redraw itself on the AxesGrid axes.
        plot = p.plots[(field_z)]
        plot.figure = fig
        #grid[i].cax.minorticks_on()
        plot.axes = grid[i].axes
        if i == 0:
            plot.cax = grid.cbar_axes[i]
            
        #grid[i].axes.yaxis.set_visible(False)
        #grid[i].axes.xaxis.set_visible(False)
        
        # Actually redraws the plot.
        p._setup_plots()
        
        # Modify the axes properties **after** p._setup_plots() so that they are not overwritten.
        #plot.axes.xaxis.set_minor_locator(plt.LogLocator(base=10.0, subs=[2.0,5.0,8.0]) )
        
        
        grid.cbar_axes[i].set_ylabel(field_z_label, fontsize=label_size)
        grid.cbar_axes[i].tick_params(labelsize = tick_size, color='k', direction='out', which='both', length=5, width=1)
        #cb.ax.minorticks_on()
        
        ### model name and time of the snapshot
        # grid[i].axes.text(-0.45, 0.4, alphabat[i] + ' ' + model_name[i], fontsize=text_size, color='w')
        #grid[i].minorticks_on()
        grid[i].axes.yaxis.tick_left()
        grid[i].axes.set_ylabel(" ", fontsize=label_size)
        grid[i].set_yticks(yticks)
        grid[i].tick_params(axis="y", labelsize=tick_size, direction='out', which='both', length=5, width=1.2)
        grid[i].tick_params(axis="y", which='minor', length=3, width=0.8)
        #grid[i].yaxis.set_minor_locator( mplb.ticker.AutoMinorLocator(10) )
        
        grid[i].axes.xaxis.tick_bottom()
        grid[i].axes.set_xlabel(" ", fontsize=label_size)
        grid[i].set_xticks(xticks)
        grid[i].tick_params(axis="x", labelsize=tick_size, direction='out', which='both', length=5, width=1.2 )
        grid[i].tick_params(axis="x", which='minor', length=3, width=0.8)
        #grid[i].xaxis.set_minor_locator( mplb.ticker.AutoMinorLocator(10) )
        
        #grid[i].axes.set_title( "({}) {}".format( chr(97+i), model_name[i]), fontsize=label_size, color='k' )
        #grid[i].axes.set_title( "{}".format(model_name[i]), fontsize=title_size, color='k' )
        
        grid[i].axes.set_title( "({}) {}".format( chr(97+i), model_name[i]), fontsize=title_size, color='k' )
        
    fig.text(0.5, 0.0075, field_x_label, ha='center', fontsize=label_size)
    fig.text(0.005, 0.5, field_y_label, va='center', rotation='vertical', fontsize=label_size)
    plt.savefig( output_dir, dpi=160 )
    
```

```python
clump_phase_diagram(Clumps, Files, output_dir_pp, ncol, nrow, figsize_x, figsize_y, label_size, tick_size, model_name,
              field_x, field_x_label, field_y, field_y_label, field_z, field_z_label,
              title_size,xticks,yticks,
              weight_field="cell_mass", colormap='PRISM')
```

```python
def _infall_rate(field, data):
    return 4*np.pi*np.power(data['index','spherical_radius'],2)*data['gas','density']*data['gas','velocity_spherical_radius']

yt.add_field(("gas", "infall_rate"), function=_infall_rate, units="Msun/yr")


datafile_names_2       = np.array([['data-Mach2to1','data-Mach2to1','data-Mach2to1','data-3rdStep-{}'.format(final_dump_2nd_2[i])] 
                                   for i in range(len(model_names_2))])

datasets = [ [yt.load("{}/test{}/{}/DD{:0>4d}/DD{:0>4d}".format(cscratch, model_names_2[i],
                                                               datafile_names_2[i,j], three_shapshots_2nd_2[i,j],three_shapshots_2nd_2[i,j]))
              for j in range(len(datafile_names_2[0]))]
            for i in range(len(datafile_names_2)) ]

nrow             = 4
ncol             = 4
title_size       = 32
ticklabel_size   = 21
xlabel_size      = 32
ylabel_size      = 30
legend_size      = 17

field_x          = 'radius'
## determine n_bins by ensuring the bin size is greater than the maximum spatial resolution
max_spatial_rsl  = 3/(rsl*2**2)
print("max_spatial_rsl =",max_spatial_rsl," pc")

field_names      = [r"Density [$\mathrm{g/cm^3}$]", r"$\mathrm{H_2}$ fraction",
                    "Temperature [K]", r"$\mathrm{R_{in-fall} \ [M_\odot/yr]}$"]
field_y          = [('gas', 'density'), ('gas', 'H2_fraction'), 
                    ('gas', 'temperature'), ('gas', 'infall_rate')]
weight_fields    = [('index', 'ones'), ('gas', 'cell_mass'), 
                    ('gas', 'cell_mass'), ('index', 'ones')]
field_units      = [{"density":"g/cm**3"}, {"H2_fraction":""}, 
                    {"temperature":"K"}, {"infall_rate":"Msun/yr"}]

ylim_rows        = [[1e-21,6e-16],[2.4e-3,4.8e-3],
                    [0,800],[1e-6,1e-1]]

xticks_cols      = [[0.03,0.06,0.09,0.12], [0.03,0.06,0.09,0.12],
                    [0.03,0.06,0.09,0.12], [0.03,0.06,0.09,0.12]]

yticks_rows      = [[1e-20,1e-19,1e-18,1e-17,1e-16], [1e-3,2e-3,3e-3,4e-3],
                    [200,400,600], [1e-8,1e-6,1e-4,1e-2]]

yticklabels_rows = [[r"$10^{-20}$",r"$10^{-19}$",r"$10^{-18}$",r"$10^{-17}$",r"$10^{-16}$"], [1e-3,2e-3,3e-3,4e-3],
                    [200,400,600], [r"$10^{-8}$",r"$10^{-6}$",r"$10^{-4}$",r"$10^{-2}$"]]

color_3shapshots_2     = ['lightsalmon','green','royalblue','k']
t_names                = [r"$\mathrm{t_1}$",r"$\mathrm{t_2}$",r"$\mathrm{t_3}$",r"$\mathrm{t_4}$"]



# Make plots using matplotlib
#fig = plt.figure(figsize=(20,20))
fig, axes = plt.subplots(nrow, ncol, figsize=(20,16))
fig.subplots_adjust(left=0.08,
                    bottom=0.06, 
                    right=0.97, 
                    top=0.97,
                    wspace=0, 
                    hspace=0)

## By using axes.T, same model plot will be in same col of axes, rather than same row of axes
## [[model1a,model1b,model1c],
##  [model2a,model2b,model2c]] 
## => 
## [[model1a,model2a],
##  [model1b,model2b],
##  [model1c,model2c]]
for l, ax_col in enumerate(axes.T): ## M4C3S, M4C4D, M8C3S, M8C4D
    print('\n\n=======================================')
    print(model_names_2[l])
    print('=======================================')
    dumps_l = datasets[l]
    
    if l!=3:
        alld = dumps_l[-1].all_data()
    else: # 0.6, 2.05, 2.4
        left_edge  = (0.15, 0.633, 0.75)
        right_edge = (0.25, 0.733, 0.85)
        alld       = dumps_l[-1].box(left_edge, right_edge)
        
    max_density       = alld["gas","density"].max()
    max_density_arg   = (alld["gas","density"]==max_density)
    print("max density =", max_density)
    max_density_pos_x = alld["index", "x"][max_density_arg][0].in_units("pc")
    max_density_pos_y = alld["index", "y"][max_density_arg][0].in_units("pc")
    max_density_pos_z = alld["index", "z"][max_density_arg][0].in_units("pc")
    print("clump max density position =", [max_density_pos_x.in_units("pc"), max_density_pos_y.in_units("pc"), max_density_pos_z.in_units("pc")])
    clump_radius_l = 0.15
    print("clump radius =", clump_radius_l)
    
    ## determine n_bins by ensuring the bin size is greater than the maximum spatial resolution
    nbins = int(float(clump_radius_l)/max_spatial_rsl)-1
    print("nbins =",nbins)
    ## to many bins will shape the lines into sawtooth
    if nbins>32: 
        nbins = 32
    ## x axis limited by the clump radius
    xlim_col         = [0, float(clump_radius_l)]
    print('=============\n')
    
    ##
    for m, ax in enumerate(ax_col): ## Density, H2 fraction, Temperature, Infall rate
        ## 
        #i = l*nrow+m
        #fig_config = nrow*100 + ncol*10 + (l*ncol+m+1)
        #print("subplt({})".format(fig_config))
        #ax = fig.add_subplot(fig_config)
        print("subplt({},{})".format(l,m))
        
        ylim_row = ylim_rows[m]
        
        ##
        for n, ds in enumerate(dumps_l): ## t1, t2, t3, t4
            print(ds)
            #alld = ds.all_data()
            
            sp = ds.sphere((float(max_density_pos_x.in_units("code_length")),
                            float(max_density_pos_y.in_units("code_length")),
                            float(max_density_pos_z.in_units("code_length"))),
                           (float(clump_radius_l), "pc"))
            
            time_ds = float(ds.current_time.in_units('Myr'))
            if n==len(dumps_l)-1:
                time_ds += final_dump_2nd_2[l]*simulation_time_step_2[l]
            
            
            # Bin up the data from the alld into a field_x profile
            profiles = yt.create_profile(
                sp, field_x, field_y[m], weight_field=weight_fields[m],
                units=field_units[m], n_bins=nbins,
                logs={"radius": False}
                #extrema={("gas", "temperature"): (1e3, 1e7), ("gas", "density"): (1e-26, 1e-22)},  
            ) 
            
            
            ## Plot the 1D profile as a semi-log plot using the default settings
            if m==0:
                ## ignore the data points that are zero for density
                arg_greaterthan0 = np.where(profiles[field_y[m]].value>0)
                profile_plot = ax.semilogy(profiles.x.value[arg_greaterthan0]/pc_cm, profiles[field_y[m]].value[arg_greaterthan0], 
                                           color_3shapshots_2[n], linestyle='-', label="{} = {:<4.3f} Myr".format(t_names[n],float(time_ds)),
                                           linewidth=3, alpha=0.5)
            elif m==3:
                ## ignore the data points that are positvie for infall rate
                """
                arg_lessthan0 = np.where(profiles[field_y[m]].value<0)
                profile_plot = ax.semilogy(profiles.x.value[arg_lessthan0]/pc_cm, -profiles[field_y[m]].value[arg_lessthan0], 
                                       color_3shapshots_2[n], linestyle='-', label="{} = {:<4.3f} Myr".format(t_names[n],float(time_ds)),
                                       linewidth=3, alpha=0.5)
                """
                ## rectify the values greater than zero to be -1e-9
                arg_greaterthan0 = np.where(profiles[field_y[m]].value>=0)
                infall_rate_arr = np.copy(profiles[field_y[m]].value)
                infall_rate_arr[arg_greaterthan0] = -1.02e-9
                profile_plot = ax.semilogy(profiles.x.value[:]/pc_cm, -infall_rate_arr[:], 
                                       color_3shapshots_2[n], linestyle='-', label="{} = {:<4.3f} Myr".format(t_names[n],float(time_ds)),
                                       linewidth=3, alpha=0.5)
                
            else:
                ## ignore the data points that are zero for H2 fraction and temperature
                arg_greaterthan0 = np.where(profiles[field_y[m]].value>0)
                profile_plot = ax.plot(profiles.x.value[arg_greaterthan0]/pc_cm, profiles[field_y[m]].value[arg_greaterthan0], 
                                       color_3shapshots_2[n], linestyle='-', label="{} = {:<4.3f} Myr".format(t_names[n],float(time_ds)),
                                       linewidth=3, alpha=0.5)
            #ax.text(1.5e-2, 2e2, "peak: {:<4.2f}".format(profiles.x.value[peak_y_arg]), fontsize=legend_size)
            
            
        ax.minorticks_on()
        
        ax.set_yticks(yticks_rows[m])
        ax.set_yticklabels([])
        if l==0:
            ax.set_yticklabels(yticklabels_rows[m])
            ax.set_ylabel("{}".format(field_names[m]), fontsize=ylabel_size)
        ##ax.yaxis.set_minor_locator( mplb.ticker.AutoMinorLocator(10) )
        
        
        ax.set_xticks(xticks_cols[l])
        ax.set_xticklabels([])
        if m==len(field_names)-1:
            ax.set_xticklabels(xticks_cols[l])
        ax.set_xlabel(" ", fontsize=xlabel_size)
        ##ax.xaxis.set_minor_locator( mplb.ticker.AutoMinorLocator(10) )
        
        ax.tick_params(axis='both',labelsize=ticklabel_size,which='both',color='k', direction='in',width=1.5, length=12)
        ax.tick_params(axis='both',which='minor',color='k', direction='in',width=0.7, length=7)
        
        if m==1:
            if l<=1:
                ax.legend(loc='upper right',fontsize=legend_size)
            if l>=2:
                ax.legend(loc='lower right',fontsize=legend_size)
        ax.set_xlim(xlim_col[0],xlim_col[1])
        ax.set_ylim(ylim_row[0],ylim_row[1])
        if m==0:
            ax.set_title('{}'.format(model_names_2[l]), fontsize=title_size)
        
        
        print('=============')

fig.text(0.5, 0.006, "Radius [pc]", ha='center', fontsize=xlabel_size)
fig.text(0.005, 0.5, " ", va='center', rotation='vertical', fontsize=ylabel_size)
#plt.tight_layout()
plt.savefig("./PlotForResult/RadiusPlot-4-noClump.png", dpi=160)
plt.show()



```
