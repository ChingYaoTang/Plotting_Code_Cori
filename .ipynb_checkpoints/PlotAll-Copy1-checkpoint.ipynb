# Scatter Plot

```python
from gravitaionally_bound_structures import *
output_dir  = ["../test2203281800", "../test2203281700", "../test2203281500",
               "../test2110121500", "../test2110121400"]
ds_clumps   = [read_clumps_info(output_dir[i]+"/clump_info.h5") for i in range(len(output_dir))]
```

```python
which_clump = [[0,1,2,3],
               [0],
               [0,1,2],
               [0,1],
               [0]]
jeans_mass       = []
angular_momentum = []
for i in range(len(output_dir)):
    jeans_mass_i       = []
    angular_momentum_i = []
    for j in which_clump[i]:
        #cc = Clumps_Info(ds_clumps[i].leaves[j])
        cc = Clumps_Info(ds_clumps[i].tree.children[j])
        jeans_mass_i.append(cc.jeans_mass)
        angular_momentum_i.append(cc.L)
    jeans_mass.append(jeans_mass_i)
    angular_momentum.append(angular_momentum_i)
```

```python
colors      = ["royalblue","aqua","black","red","greenyellow"]
names       = ["M8C5D","M4C5D","M8C5S","M8C4S","M4C4S"]
plt.figure(figsize=(8, 8), dpi=100)
for i in range(len(output_dir)):
    plt.scatter(jeans_mass[i], angular_momentum[i], c=colors[i],s=200, label=names[i])
plt.xlabel(r"mass-weighted Jeans mass ($M_\odot$)", fontdict={'size': 16})
plt.ylabel(r"mass-weighted angular momentum ($g \cdot cm^2 / s$)", fontdict={'size': 16})
plt.legend(loc='best',fontsize=16)
plt.tight_layout()
plt.savefig("Mj-L-")
plt.show()
```

# Density projection plot in 3 simulation stages

```python
model_name_three_snapshot = ['M4C2D', 'M4C4D', 'M8C2D', 'M8C4D']

three_snapshot = np.array([ 
    model_name_three_snapshot[0] + '/data/DD0060/DD0060',              # End of the 1st stage simulation
    model_name_three_snapshot[0] + '/data-Mach2to1/DD0082/DD0082',     # After 1 turnover time in Mach2to1 in 2nd stage simulation
    model_name_three_snapshot[0] + '/data-3rdStep-167/DD0017/DD0017',  # End of the 3rd stage simulation
    
    model_name_three_snapshot[1] + '/data/DD0076/DD0076',
    model_name_three_snapshot[1] + '/data-Mach2to1/DD0106/DD0106',
    model_name_three_snapshot[1] + '/data-3rdStep-148/DD0017/DD0017',
    
    model_name_three_snapshot[2] + '/data/DD0033/DD0033',
    model_name_three_snapshot[2] + '/data-Mach2to1/DD0066/DD0066',
    model_name_three_snapshot[2] + '/data-3rdStep-210/DD0017/DD0017',
    
    model_name_three_snapshot[3] + '/data/DD0032/DD0032',
    model_name_three_snapshot[3] + '/data-Mach2to1/DD0075/DD0075',
    model_name_three_snapshot[3] + '/data-3rdStep-122/DD0004/DD0004',
])

final_dump_2nd_stage      = np.array([167, 148, 210, 122])
simulation_time_step      = np.array([0.1, 0.1, 0.05, 0.05]) # 0.1Myr
final_snaposhot_2nd_stage = final_dump_2nd_stage*simulation_time_step*0.1
```

```python
ds = [yt.load("{}/test{}".format(cscratch, three_snapshot[i])) for i in range(len(three_snapshot))]

field_name       = 'density'
bounds           = (3e0, 3e-3)
cmap_name        = 'jet'
ncol             = int(len(three_snapshot)/len(model_name_three_snapshot))
nrow             = len(model_name_three_snapshot)
title_size       = 20
ticklabel_size   = 18
xlabel_size      = 28
ylabel_size      = 28
zlabel_size      = 30
zticklabel_size  = 20
color_model_name = ['red','red','blue','blue']


fig = plt.figure(figsize=(24,18))

# See http://matplotlib.org/mpl_toolkits/axes_grid/api/axes_grid_api.html
# These choices of keyword arguments produce a four panel plot with a single
# shared narrow colorbar on the right hand side of the multipanel plot. Axes
# labels are drawn for all plots since we're slicing along different directions
# for each plot.

grid = AxesGrid(fig, (0.01,0.06,0.98,0.92),
                nrows_ncols = (nrow, ncol),
                axes_pad = 0.4,
                label_mode = "L",
                share_all = True,
                cbar_location="right",
                cbar_mode="single",
                cbar_size="3%",
                cbar_pad="2%")

for i in range(len(ds)):
    print(ds[i])
    alld = ds[i].all_data()
    time_ds = ds[i].current_time.in_units('Myr')
    # Load the data and create a single plot
    p = yt.ProjectionPlot(ds[i], 'z', field_name, center='c')

    # Ensure the colorbar limits match for all plots
    p.set_zlim(field_name, bounds[0], bounds[1])
    
    p.set_font({'family': 'sans-serif', 'style': 'normal',
              'weight': 'book', 'size': 30})
    
    p.set_figure_size(15)
    #p.annotate_title('({}) {} Myrs'.format(chr(97+i),time_ds))
    p.set_cmap(field=field_name, cmap=cmap_name)
    p.set_cbar_minorticks('all',True)
    # This forces the ProjectionPlot to redraw itself on the AxesGrid axes.
    plot = p.plots[field_name]
    plot.figure = fig
    grid[i].cax.minorticks_on()
    plot.axes = grid[i].axes
    plot.cax = grid.cbar_axes[i]
    
    grid[i].axes.yaxis.set_visible(False)
    grid[i].axes.xaxis.set_visible(False)
    # Finally, this actually redraws the plot.
    p._setup_plots()

    
    ### colorbar
    grid.cbar_axes[i].set_ylabel(r"Projection density [$\mathrm{g/cm^2}$]", fontsize=zlabel_size)
    grid.cbar_axes[i].tick_params(labelsize = zticklabel_size, which='both',
                                  color='k', direction='out',
                                  width=2, length=10)
    grid.cbar_axes[i].tick_params(which='minor', width=1, length=7)

    ### ticks, ticklabels, label, model name and time
    if i%ncol == 0:
        grid[i].axes.yaxis.set_visible(True)
        grid[i].axes.yaxis.tick_left()
        ##
        grid[i].axes.set_ylabel('{}'.format(model_name_three_snapshot[int(i/ncol)]), fontsize=ylabel_size, color=color_model_name[int(i/ncol)])
        ## 
        grid[i].tick_params(axis="y", labelsize=ticklabel_size, direction='out', which='both', width=2, length=7)
        grid[i].tick_params(axis="y", which='minor', width=1, length=3)
        
    if i>=(nrow-1)*ncol:
        grid[i].axes.xaxis.set_visible(True)
        grid[i].axes.xaxis.tick_bottom()
        ##
        grid[i].axes.set_xlabel('')
        ##
        grid[i].tick_params(axis="x", labelsize=ticklabel_size, direction='out', which='both', width=2, length=7)
        grid[i].tick_params(axis="x", which='minor', width=1, length=3)
    
    grid[i].set_yticks([-1.2, -0.6, 0, 0.6, 1.2])
    grid[i].set_xticks([-1.2, -0.6, 0, 0.6, 1.2])
    
    if i%ncol==2:
        grid[i].axes.set_title('({}{}) {:<4.3f} Myr'.format(chr(97+int(i/ncol)), int(i%ncol)+1,
                                                            float(time_ds)+final_snaposhot_2nd_stage[int(i/ncol)]),
                               fontsize=title_size, color=color_model_name[int(i/ncol)])
    else:
        grid[i].axes.set_title('({}{}) {:<4.3f} Myr'.format(chr(97+int(i/ncol)), int(i%ncol)+1,
                                                            float(time_ds)), 
                               fontsize=title_size, color=color_model_name[int(i/ncol)])

fig.text(0.5, 0.01, 'x [pc]', ha='center', fontsize=xlabel_size)
fig.text(0.08, 0.5, 'y [pc]', va='center', rotation='vertical', fontsize=ylabel_size)
plt.savefig('./PlotForResult/projection_density_re.png', dpi=360)
```
